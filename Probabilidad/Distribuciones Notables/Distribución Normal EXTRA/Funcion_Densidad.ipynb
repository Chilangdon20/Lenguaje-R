# Funciones de densidad de Probabilidad.

Hasta ahora hemosa hablado de probabilidades de cantidades discretas, como tirar un dado, número de llegadas a un autobus, pero que pasa con las cantidades que son continuas.

Una cantidad continua puede tomar cualquier valor , no solo discreto, por ejemplo la velocidad de un carro es de 145.3km /hora.

Estas variables también tienen distribuciones de probabilidad, veamos un ejemplo:


En 1879,Alber Michelson realizó 100 mediciones de la velocidad de la luz en el aire.Cada medida tiene algún error,condiciones como la temperatura, humedad, alineacion de su óptica etc, cambio de medida en medida.

Como resultado , es posible cualquier valor fraccional de la velocidad medida de la luz.

Estos resultados son aptos para describirlos con una distribución de probabilidad continua.
