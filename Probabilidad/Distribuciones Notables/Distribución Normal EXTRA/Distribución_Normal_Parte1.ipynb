# Introducción a la distribución Normal.

La distribución normal es famosa y la usamos como ejemplo para aprender sobre distribuciones continuas.

No entraremos en la historia de esta distribucion en profundidad, es suficiente decir que describe una variable continua cuyo PDF es simétrico y tiene un pico único & asi es como se ve:

![texto alternativo](https://www.boost.org/doc/libs/1_39_0/libs/math/doc/sf_and_dist/graphs/pdf.png)

Esta distribución esta parametrizada por dos parámetros, la média que determina donde esta el centro del pico & la desviacion estandar que es una medida que describe que tan ancho es el pico o que tan dispersos están los datos.

La média y la desviación estandar o std son los nombres de los dos parámetros de la distrbucion Normal.

No confundamos estos con la media y la std que nosotros calculamos directamente apartir de los datos al momento de hacer un análisis exploratorio.

![texto alternativo](https://raw.githubusercontent.com/Chilangdon20/IMAGENES/master/Imagenes%20probabilidad/Distirbuci%C3%B3n%20Normal/Captura%20de%20Pantalla%202020-08-07%20a%20la(s)%209.51.18.png?token=AOSXAEDORQP3ZDN2ZM64XJK7G2SXQ)

Ahora , si agregamos un histograma a las medidas de Michelson , vemos que la velocidad de la luz en el aire parece estar normalmente distribuida.

La comparación del histograma con el PDF tiene sesgo de agrupamiento, por lo que es mejor comparar el ECDF de los datos con el CDF teórico de la distribución normal.

Para calcular el CDF teórico, podemos usar `np.random.normal()` para extraer muestras y luego podemos calcular el CDF.

A continuación veremos algunos ejemplos:

Exploraremos el PDF normal y también aprenderemos una forma de trazar un PDF de una distribución conocida utilizando estadísticas de piratas informáticos. Específicamente, trazará un PDF normal para varios valores de la varianza.

PASOS:

* Extraemos 100.000 muestras de una distribución normal que tenga una media de 20 y una desviación estándar de 1. Hagamos lo mismo para las distribuciones normales con desviaciones estándar de 3 y 10, cada una con una media de 20. Asignemos los resultados a `samples_std1`, `samples_std3` y `samples_std10`, respectivamente.

* Trazamos un histograma de cada una de las muestras; para cada uno, usamos 100 bins, también usando los argumentos de palabras clave `normed=True` e `histtype = 'step'`. El último argumento de palabra clave hace que la trama se parezca mucho al PDF teórico fluido.

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Estilo del gráfico
sns.set()
# Extraer 100,000 muestras de la distribución normal con std de interés: samples_std1, samples_std3, samples_std10
samples_std1 = np.random.normal(20, 4, size=100000)
samples_std3 = np.random.normal(20, 1, size=100000)
samples_std10 = np.random.normal(20, 10, size=100000)

# Trazar histogramas
_ = plt.hist(samples_std1, bins=100, density=True, histtype='step',color='orange')
_ = plt.hist(samples_std3, bins=100, density=True, histtype='step',color='red')
_ = plt.hist(samples_std10, bins=100, density=True, histtype='step',color='blue')
_ = plt.xlabel("x")
_ = plt.ylabel("y")
# Definir leyenda y limites del grafico
_ = plt.legend(('std = 4', 'std = 1', 'std = 10'))
plt.ylim(-0.01, 0.42)
plt.show()
```

# El CDF normal

Ahora que tenmos  una idea de cómo se ve el PDF normal, consideremos su CDF. Con las muestras que generamos anteriormente (en su espacio de nombres como samples_std1, samples_std3 y samples_std10), generemos y grafiquemos los CDF.

```
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y
```

```
# Generar ECDF
x_std1, y_std1 = ecdf(samples_std1)
x_std3, y_std3 = ecdf(samples_std3)
x_std10, y_std10 = ecdf(samples_std10)

# Plot
_ = plt.plot(x_std1, y_std1, marker='.', linestyle='none',color='orange')
_ = plt.plot(x_std3, y_std3, marker='.', linestyle='none',color='red')
_ = plt.plot(x_std10, y_std10, marker='.', linestyle='none',color='blue')
_ = plt.xlabel("x")
_ = plt.ylabel("y")
# Leyenda y mostrar plot
_ = plt.legend(('std = 4', 'std = 1', 'std = 10'), loc='lower right')
plt.show()
```
