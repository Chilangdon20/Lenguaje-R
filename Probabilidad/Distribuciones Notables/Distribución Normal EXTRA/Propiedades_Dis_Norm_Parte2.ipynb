# Propiedades de la distribución Normal.

La distribución normal amenudo también se le llama distribución gaussiana , en honor a su inventor , por lo que a menudo escucharas que se hace referencia de esta manera.


Entonces la distribución normal y es muy importante y muy utilizada, es tan frecuente que vale la pena hablar un poco más de ella.

En la practica se utiliza para describir la mayoría de los datos pico simétricos que encontraremos como cientificos de datos.

Ademas , para muchos de los procedimientos estadísticos , nosotros hemos escuchada hablar sobre la normalización de datos.De hecho es una distribución que parece ser de naturaleza [ubicua](https://es.thefreedictionary.com/ubicua), no solo en el campo de la estadística.

Dicho esto, hay advertencias sobre la distribución y debemos de tener cuidado al usarla.

* En primer lugar , muchas veces las cosas que podemos pensar que están distribuidad de forma normal no lo están.

Consideremos por ejemplo , el bajo de boca grande , Lagos de Massachusstes medidos en 1994 y 1995 por el Departamento de protección ambiental de Massachussetts.
