## EJEMPLOS-(Relaciones exponenciales & potenciales.)

```r
dep = c(1.2,3.6,12,36)
ind = c(20,35,61,82)
plot(ind,dep, main="Escala Lineal")
plot(ind,dep,log='y',main="Escala Semiogaritmica")
```


Podemos ver que una vez realizada la transformacion logaritmica al eje de las y la grafica se puede ver como una recta , una vez hecho esto procedemos a hacer el modelo linea..

```r
regresion = lm(log10(dep)~ind)
regresion
summary(regresion)$r.squared
```

Realizamos un plot de la curva de regresion en escala lineal.

```r
plot(ind,dep, main="Curva de regresion")
curve(1.054^x*0.468, add=TRUE, col="red")
```

## Modelo Potencial.

```r
tiempo = 1:10
tarifa = c(0.097,0.709,2.698,6.928,15.242,29.994,52.902,83.903,120.612,161.711) 
data = data.frame(tiempo,tarifa)
```

Hacemos  los graficos correspondientes.

```r
plot(data)
plot(data,log="y",main="Grafico Semilogaritmico")
plot(data,log="xy",main="Grafico DobleLogaritmico")
```

Aplicamos una transformacion logaritmica.

```r
transformacion = lm(log10(tarifa)~log10(tiempo),data=data)
transformacion
summary(transformacion)$r.squared
```

Graficamos la curva de regresion:

```r
plot(data,main="Curva de regresion")
curve(x^(3.298)*0.081, add=TRUE, col="Blue")
```
