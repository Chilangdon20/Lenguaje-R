
## Distribución Binomial

Si $X$ es variable aleatoria que mide el "número de éxitos" y se realizan $n$ ensayos de Bernoulli independientes entre sí, diremos que $X$ se distribuye como una Binomial con parámetros $n$ y $p$

$$X\sim \text{B}(n,p)$$

donde $p$ es la probabilidad de éxito y $q = 1-p$ es la probabilidad de fracaso

- El **dominio** de $X$ será $D_X = \{0,1,2,\dots,n\}$
- La **función de probabilidad** vendrá dada por $$f(k) = {n\choose k}p^k(1-p)^{n-k} $$

## Distribución Binomial

- La **función de distribución** vendrá dada por $$F(x) = \left\{
\begin{array}{cl}
     0 & \text{si } x<0 
  \\ \sum_{k=0}^xf(k) & \text{si } 0\le x<n
  \\ 1 & \text{si } x\ge n
\end{array}
\right.$$
- **Esperanza** $E(X) = np$
- **Varianza** $Var(X) = npq$

<l class = "important">Atención.</l> Fijemonos que la distribución de Bernoulli es un caso particular de la Binomial. Basta tomar $n=1$ y tendremos que $X\sim \text{Be}(p)$ y $X\sim\text{B}(1,p)$ son equivalentes.

```python
par(mfrow = c(1,2))
plot(0:50,dbinom(0:50,50,0.5),col = "red", xlab = "No experimeto", ylab = "probabilidad", main = "F.D.P de una B(50,0.5)")
plot(0:50, pbinom(0:50,50,0.5),col = "purple", xlab = "Num experimento", ylab = "probabilidad", main = "F.D de una B(50,0.5)", ylim = c(0,1))
par(mfrow= c(1,1))
```

## El código de la distribución Binomial:

- En **`R`** tenemos las funciones del paquete `Rlab`: 
* `dbinom(x, size, prob)

*  pbinom(q,size, prob)

* qbinom(p, size, prob)

* rbinom(n, size, prob)

* ` donde `prob` es la probabilidad de éxito y `size` el número de ensayos del experimento.


- En **`Python`** tenemos las funciones del paquete `scipy.stats.binom`: 

* `pmf(k,n,p)
*  cdf(k,n,p)
*  ppf(q,n,p)
*  rvs(n, p, size)`
*  donde `p` es la probabilidad de éxito y `n` el número de ensayos del experimento.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1,1)

n = 10
p = 0.3

mean,var,skew,kurt = binom.stats(n,p,moments = 'mvsk')

x = np.arange(0,n+1)
ax.plot(x,binom.cdf(x,n,p),'bo',ms=8,label="funcion de densidad de una binomial")
ax.vlines(x,0,binom.cdf(x,n,p),colors = 'orange', lw=4, alpha=0.5)
rv = binom(n,p)
ax.vlines(x,0,rv.pmf(x),colors="black",linestyles='--',lw=1,label="Distribucion teorica")
ax.legend(loc = 'best',frameon=False)
plt.show()

v = binom.rvs(n,p,size =10000)
```
